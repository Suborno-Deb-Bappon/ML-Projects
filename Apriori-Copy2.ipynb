```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
store_data = pd.read_csv('1604001 Depression Analysis.csv',head=None)
```

```python
len(store_data.columns)
len(store_data)
```

```python
store_data.head()
```

```python
records = []
for i in range(0, 1005):
    records.append([str(store_data.values[i, j]) for j in range(0, 21)])
```

```python
association_rules = apriori(records, min_support=0.0765, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
print(len(association_results))
print(association_results[0])
```

```python
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
