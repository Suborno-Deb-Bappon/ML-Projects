```python
import pandas as pd 
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data=pd.read_csv('DHF.csv')
data_new=data.iloc[:, 1:].copy()
```

```python
info=data_new .to_numpy().tolist()
```

```python
for values in info:
    for i in range(len(values)):
        s= ''
        if type(values[i])!=str:
            s=str(values[i])
            values[i]=s
        
```

```python
tr= TransactionEncoder()
```

```python
t_info=tr.fit(info).transform(info)
```

```python
df=pd.DataFrame(t_info, columns=tr.columns_)
```

```python
frequent_items=apriori(df,min_support=0.60, use_colnames=True)
print(frequent_items)
```

```python
res=association_rules(frequent_items,metric='confidence',min_threshold=0.7)
res
```
